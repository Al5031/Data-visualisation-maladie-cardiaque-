```python
## Importer les données 
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_table("coeur.txt",sep="\t")
```

```python
df
```

###  l’histogramme associé à l’âge des patient

```python
df['age'].plot(kind='hist',title="Age des patients",color="b")
plt.show()
```

###  l’âge des patients selon le sexe

```python
df.hist(column = "age", by = "sexe")
```

```python
plt.hist2d(df['age'], df['taux_max']) 
plt.colorbar() 
```

### les boîtes à moustache associée à vos données.


```python
df.plot(kind="box")
```

### Nuage des points 

```python
df.plot("age","taux_max",kind="scatter",alpha=0.9)
```

### Nuage des points (taux-max,age,depression)

```python
import matplotlib.cm as cm 
df.plot.scatter(x='age',y='taux_max',c='depression', cmap=cm.plasma_r) 
```

### Detecter la présence de la maladie cardiaque

```python
codage = df["coeur"].eq('presence').astype('int') 
df.plot.scatter(x='age',y='taux_max',c=pd.Series(['blue','red'])[codage]) 
```
